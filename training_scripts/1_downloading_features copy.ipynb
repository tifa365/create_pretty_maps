# Downloading map features with OpenStreetMap


## Import of necessary libraries


```python
# Import osmnx package to download the OSM data
import osmnx as ox

# Import geopandas to handle the geographic data
import geopandas as gpd

# Import the visualiyation package matplotlib to plot the data later
import matplotlib.pyplot as plt

# Import pandas in case we have to do any other data manipulation
import pandas as pd

# set max number of columns and width in this notebook
pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_columns", None)
```

## Using the osmnx python library to download a single OpenStreetmap feature

OSMnx is a Python package to easily download, model, analyze, and visualize geospatial features from OpenStreetMap. The python package is abstracting away a lot of the complexity of an Overpass API call to download OSM data. That makes the package quite easy to use. [As an example, we will the download outline of the hospital called "Klinikum Magdeburg."](https://www.openstreetmap.org/way/25939200)

In case you do not make any progress wit, [this tutorial by Geoff Boeing](https://geoffboeing.com/2016/11/osmnx-python-street-networks/) is quite helpful.


```python
# Geocode the the hospital, btw. "ox" is short for "osmnx"
hospital = ox.geocode_to_gdf("Klinikum Magdeburg, Magdeburg, Germany")

# Plot the hospital
ax = ox.project_gdf(hospital).plot()

# Turn off the axis - and try to turn it back on later!
_ = ax.axis("off")
```

Congratulations, you have downloaded and already visualized your first feature! If you still have some time to spare, you can continue on your own, for example, geocode a feature from your current place of living or home town, or continue with one of the following tasks.

## Coloring the feature differently

On a map, similar entities are typically represented by the same color. The Python visualization library matplotlib provides us with the flexibility to customize this color scheme using the `color` parameter during the plotting process. Let's proceed to assign a red color to the hospital to distinguish it from the default blue.


```python
# Plot the hospital with red color
ax = ox.project_gdf(hospital).plot(color="red")

# Turn off the axis
_ = ax.axis("off")
```

## Coloring the feature and the map's background

Well, that was a bit too easy, wasn't it? To be frank, the base map (the background all features are placed upon) looks a bit unspectactular. On a real map, it would perhaps be grey but definitely not white.


```python
# Retrieve and project the hospital data
hospital_gdf = ox.geocode_to_gdf("Klinikum Magdeburg, Magdeburg, Germany")
hospital_proj = ox.project_gdf(hospital_gdf)

# Create a base map to plot against, for context
city_gdf = ox.geocode_to_gdf("Magdeburg, Germany")
city_proj = ox.project_gdf(city_gdf)

# Plot the city as the base layer, the edge colors could be also called the border color
ax = city_proj.plot(figsize=(10, 10), color="grey", edgecolor="black")

# Plot the hospital on top with red color
hospital_proj.plot(ax=ax, color="red")

# Remove the axis for a cleaner look
ax.axis("off")

# Display the plot
plt.show()
```
